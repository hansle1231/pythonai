```python
from konlpy.tag import Okt  # okt=한국어 형태소 분석
```

```python
readFp=codecs.open(r'C:\Users\Affinity\aaaa\kowiki-latest-pages-articles.xml\wiki.wp2txt',encoding='utf-16')
gubun_file='wiki.gubun'


writeFp=open(gubun_file,'w',encoding='utf-16')

okt=Okt()

i=0

while True:
    line=readFp.readline()
    if not line:break
    if i%20000==0:
        print('current-'+str(i))
    i+=1
    malist=okt.pos(line,norm=True,stem=True)
    r=[]
    
    for word in malist:              
        if not word[1] in ['josa','Eomi','Punctuation']:  #명사만 걸러줌
            writeFp.write(word[0]+'')
writeFp.close()
```

```python
from gensim.models import word2vec
sentences=word2vec.Text8Corpus('./wiki.gubun')  #목록집을 만듬  
model=word2vec.Word2Vec(sentences,sg=1,size=100,window=5)
model.save('./wiki.model')
                            
```

```python
from gensim.models import word2vec
model=word2vec.Word2Vec.load('./wiki.model')
results=model.wv.most_similar(positive=['과자'])
for result in results:
    print(result)
```

```python
#시급성 판단  #결과들의 평균으로 얼마나 시급한지 알려준다.
model=word2vec.Word2Vec.load('./wiki.model')
def print_emargency(text):
    print(text)
    node=okt.pos(text,norm=True,stem=True)
    for word,form in node:
        if form=='Noun' or form=='Verb' or form=='Adgective' or form=='Adverb':
            
            print('-',word,':',model.wv.similarity(word,'급하다'))
print_emargency('도와주세요. 문제가 생겼습니다. 환자가 있어요')
print_emargency('사용방법을 잘 모르겠습니다.')
```

```python
import codecs
from bs4 import BeautifulSoup
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.layers import LSTM
from keras.optimizers import RMSprop
from keras.utils.data_utils import get_file
import numpy as np
import random,sys
fp=codecs.open(r'C:\Users\Affinity\aaaa\BEXX0003.txt',encoding='utf-16') #토지 데이터를 이용한다.
```

```python
fp=codecs.open(r'C:\Users\Affinity\aaaa\BEXX0003.txt',encoding='utf-16')
soup=BeautifulSoup(fp,'html.parser')
body=soup.select_one('body')
text=body.getText()+' '


print('코퍼스의 길이:',len(text))
chars=sorted(list(set(text)))
print('사용되고있는 문자의수:',len(chars))
char_indices=dict((c,i) for i,c in enumerate(chars))  #i 숫자 ,c캐릭터
indices_char=dict((i,c) for i,c in enumerate(chars))

```

```python
maxlen=20
step=3
sentences=[]
next_chars=[]
for i in range(0,len(text)-maxlen,step):
    sentences.append(text[i:i+maxlen])
    next_chars.append(text[i+maxlen])
X=np.zeros((len(sentences),maxlen,len(chars)),dtype=np.bool)
y=np.zeros((len(sentences),len(chars)),dtype=np.bool)
print('학습할 구문수',len(sentences))
print('텍스트를 ID 벡터로 변환합니다.')
    
for i, sentence in enumerate(sentences):
    for t, char in enumerate(sentence):
        X[i,t,char_indices[char]]=1
    y[i,char_indices[next_chars[i]]]=1
    
```
