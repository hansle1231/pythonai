```python
class prepro:
    
    from hgtk.letter import compose,decompose
    
    def __init__(self,text,delete_bracket=False):
        from copy import deepcopy
        from functools import reduce
        from hgtk import checker
        from hgtk.letter import compose,decompose 
        from eunjeon import Mecab
        self.idx=[]
        sixteen_dict={'뒷윷':'뒨뉻','도리깻열':'도리깬녈','대팻밥':'대팯빱','디귿이':'디그시','디귿을':'디그슬','디귿에':'디그세','지읒이':'지으시','지읒을':'지으슬','지읒에':'지으세','치읓이':'치으시','치읓을':'치으슬','치읓에':'치으세','키읔이':'키으기','키읔을':'키으글','키읔에':'키으게','티읕이':'티으시','티읕을':'티으슬','티읕에':'티으세','피읖이':'피으비','피읖을':'피으블','피읖에':'피으베','히읗이':'히으시','히읗을':'히으슬','히읗에':'히으세'}
        for i in sixteen_dict.keys():
            if i in text:
                text.replace(i,sixteen_dict[i])
        
        # for i in range(len(text)):
        #     if text[i]==' ':
        #         self.idx.append(i)
        
        
        self.origin_text=text   #원문 비교용
        self.qq=deepcopy(text)   # 처리할 텍스트

        # 괄호안 제거해주기
        
        if delete_bracket==True:
            a=0
            b=0
            flag=0
            i=0
            while(flag==0):

                if i==len(self.qq):
                    break

                if self.qq[i]== '(':
                    a=i
                if self.qq[i]==')':
                    b=i+1
                if  (b!=0):

                    self.qq=self.qq.replace(self.qq[a:b],'')
                    a=0
                    b=0
                    i=0

                i+=1     



        
        
        self.ww=[]  #[ (비한글인덱스, 비한글)]
        
        #한글 아닌놈들 인덱스랑 값 저장
        for i in range(len(self.qq)):
            if not checker.is_hangul(self.qq[i]):
        
                    
                    self.ww.append([i,self.qq[i]])
        for i in self.ww:
            if i[1]==' ':
                continue
            self.qq=self.qq.replace(i[1],'')
        
        self.ww.sort()
        
        mecab=Mecab()
        
        self.parse=mecab.tagger.parse(self.qq)
        
        self.text=[]
        
        for i in self.parse.split('\n'):
            
            if 'Compound' in i:
                a=i.split(",")[0].split('\t')+['Compound']
                self.text.append(a)
    
            else:
                a=i.split(",")[0].split('\t')
                self.text.append(a)
            
        
        del_list=[]
        for n,i in enumerate(self.text):
            if len(i)==1:
                del_list.append(i)            
            else:    
                self.text[n][1]=i[1].split('+')[0]
      
        for i in del_list:
            self.text.remove(i)
        
        
        del_list=[]
        for n,i in enumerate(self.text): 
            if i[1][0]=='S':
             
                del_list.append(i)

        for i in del_list:
            self.text.remove(i)
        
      
  
            
        self.listed_text=[]
        for i in self.text:
            self.listed_text.append(list(i))
        for i in range(len(self.listed_text)):
            q=[]
            if self.listed_text[i][1] not in ['SN','SF','SL']:
                
                if ' ' in self.listed_text[i][0]:
                    self.listed_text[i][0]=self.listed_text[i][0].replace(' ','')
                    
                for r in self.listed_text[i][0]:
                   
                    try:
                        q.append(list(decompose(r)))
                        
                    except:
                        q.append(r)
            else:
                q.append([self.listed_text[i][0]])
            self.listed_text[i][0]=q
    
    pronounce_dict={'ㅇ':'ㅇ','ㄲ':'ㄱ','ㅋ':'ㄱ','ㅅ':'ㄷ','ㅆ':'ㄷ','ㅈ':'ㄷ','ㅊ':'ㄷ','ㅌ':'ㄷ','ㅍ':'ㅂ','ㄳ':'ㄱ', 'ㄵ':'ㄴ', 'ㄼ':'ㄹ', 'ㄽ':'ㄹ','ㄾ':'ㄹ', 'ㅄ':'ㅂ','ㄷ':'ㄷ','ㄱ':'ㄱ','ㄴ':'ㄴ','ㄹ':'ㄹ','ㅂ':'ㅂ'}

    dsr_dict={'ㄱ':'ㄲ', 'ㄷ':'ㄸ', 'ㅂ':'ㅃ', 'ㅅ':'ㅆ', 'ㅈ':'ㅉ'}
    
    
    def five(self):
        from copy import deepcopy
        text=deepcopy(self.listed_text)
        
        if len(text)>1:
            for i in range(len(text)-1):
                if text[i][1]=='VV' and text[i+1][1]=='EC':
                    if text[i][0][-1][0] in ['ㅈ','ㅊ','ㅉ'] and text[i][0][-1][1]=='ㅣ'  and text[i][0][-1][2]=='':
                        if text[i+1][0][0][0]=='ㅇ' and text[i+1][0][0][1]=='ㅓ':
                            text[i][0][-1][1]='ㅓ'
                            text[i+1][0][0]=['','','']
        
        for i in range(len(text)):
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    
                    if text[i][0][q][1]=='ㅢ' and text[i][0][q][0]!='ㅇ':
                        text[i][0][0][1]='ㅣ'
        
        return text
      
      
    def nine(self):
        
        text=self.five()
        
        nine_dict={"ㄲ":'ㄱ',"ㅋ":'ㄱ','ㅅ':'ㄷ','ㅆ':'ㄷ','ㅈ':'ㄷ','ㅌ':'ㄷ','ㅊ':'ㄷ','ㅍ':'ㅂ'}
        if len(text)==1 and len(text[0][0])==1:
            if text[0][0][0][2] in nine_dict.keys():
                return nine_dict[text[0][0][0][2]]
        
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            #형태소 내
            if text[i][1][0] in ['N','V','X'] and len(text[i][0])>1:
                  
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in nine_dict.keys() and text[i][0][q+1][0]!='ㅇ':
                        text[i][0][q][2]=nine_dict[text[i][0][q][2]]
            
            #동사일경우 -어말앞
            if len(text)>1:    
                if text[i][1][0] in ['E','V','N','X'] and text[i+1][1][0] in ['E','N'] :
                    
                    if text[i][0][-1][2] in nine_dict.keys() and text[i+1][0][0][0]!='ㅇ':
                        text[i][0][-1][2]=nine_dict[text[i][0][-1][2]]
                   
            if text[i][1]=='XPN' and text[i][0][0][2] in nine_dict.keys():
                text[i][0][0][2]=nine_dict[text[i][0][0][2]]
       
        if text[-1][0][-1][2] in nine_dict.keys():
            
            text[-1][0][-1][2]=nine_dict[text[-1][0][-1][2]]
            
        return text
    
    def ten(self):
        ten_dict={'ㄳ':'ㄱ', 'ㄵ':'ㄴ', 'ㄼ':'ㄹ', 'ㄽ':'ㄹ','ㄾ':'ㄹ', 'ㅄ':'ㅂ'}
        text=self.nine()
        if len(text)==1 and len(text[0][0])==1:
            if text[0][0][0][2] in ten_dict.keys():
                return ten_dict[text[0][0][0][2]]
              
              
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            if len(text)>1:
            #밟 예외
                if text[i][0][0]==['ㅂ','ㅏ','ㄼ'] and text[i+1][0][0][0] not in ['ㅇ']:
                    text[i][0][0]=['ㅂ','ㅏ','ㅂ']
                    continue
            #넓죽/넓둥 예외
            if len(text[i][0])>2:
               
                if [text[i][0][0],text[i][0][1]] == [['ㄴ', 'ㅓ', 'ㄼ'],['ㅈ','ㅜ','ㄱ']]:
                    text[i][0][0] = ['ㄴ','ㅓ','ㅂ']
                    text[i][0][1] = ['ㅉ','ㅜ','ㄱ']

                    continue

                if [text[i][0][0],text[i][0][1]]==[['ㄴ','ㅓ','ㄼ'],['ㄷ','ㅜ','ㅇ']]:
                    [text[i][0][0],text[i][0][1]]=[['ㄴ','ㅓ','ㅂ'],['ㄸ','ㅜ','ㅇ']]
                    continue

                #명사일경우
                if text[i][1][0]=='N':

                    for q in range(len(text[i][0])):
                        if text[i][0][q][2] in ten_dict.keys() and text[i+1][0][0][0] not in ['ㅇ']:
                            text[i][0][q][2]=ten_dict[text[i][0][q][2]]
            
            
            #24항 선적용
            if text[i][1][0]=='V' and text[i][0][-1][2] in ['ㄼ', 'ㄾ']:
                   
                    if text[i+1][1][0]=='E' and text[i+1][0][0][0] in ['ㄱ', 'ㄷ', 'ㅅ', 'ㅈ']:
                        text[i+1][0][0][0]=self.dsr_dict[text[i+1][0][0][0]]
                
            if text[i][1][0] in ['N','V'] and len(text[i][0])>1:
                  
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in ten_dict.keys() and text[i][0][q+1][0]!='ㅇ':
                        text[i][0][q][2]=ten_dict[text[i][0][q][2]]
            
            #동사일경우 -어말앞
            if len(text)>1:    
                if text[i][1][0] in ['V','N'] and text[i+1][1][0] in ['E','N'] :
                
                    if text[i][0][-1][2] in ten_dict.keys() and text[i+1][0][0][0]!='ㅇ':
                        text[i][0][-1][2]=ten_dict[text[i][0][-1][2]]
                   
            if text[i][1]=='XPN' and text[i][0][0][2] in ten_dict.keys():
                text[i][0][0][2]=ten_dict[text[i][0][0][2]]
        return text
    
    
    
    
    
    
    def eleven(self):
        
        text=self.ten()
        
        elevendict={'ㄺ':'ㄱ', 'ㄻ':'ㅁ', 'ㄿ':'ㅂ'}
        
        if len(text)==1 and len(text[0][0])==1:
            if text[0][0][0][2] in elevendict.keys():
                text[0][0][0][2]=elevendict[text[0][0][0][2]]
                
                return text
        
        if len(text)==1:
            
            
            
            l=1
        else:
            l=len(text)-1
        
        
        for i in range(l):
            #맑게 맑고
            if len(text)>1:
                if text[i][1] in ['NNG','NP','NR','VV','VA','VX'] and text[i+1][1][0] =='E' :
                    if text[i][0][0][2]=='ㄺ' and text[i+1][0][0][0]=='ㄱ':
                      text[i][0][0][2]='ㄹ'
          
            #형태소내 변화
            
            if text[i][1][0] in ['N','V'] and len(text[i][0])>1:
                if text[i][1]=='VV' :
                    continue
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in elevendict.keys() and text[i][0][q+1][0]!='ㅇ':
                        text[i][0][q][2]=elevendict[text[i][0][q][2]]
            
            #동사일경우 -어말앞
            if len(text)>1:    
                if text[i][1][0] in ['V','N'] and text[i+1][1][0] in ['E','N'] :
                
                    if text[i][0][-1][2] in elevendict.keys() and text[i+1][0][0][0]!='ㅇ':
                        text[i][0][-1][2]=elevendict[text[i][0][-1][2]]
        
             
        
        
        
        
        return text
    
    def twelve(self):
        split_dict={'ㄹ':('','ㄹ'),'ㅎ':('','ㅎ'),'ㄶ':('ㄴ','ㅎ'),'ㅀ':('ㄹ','ㅎ'),'ㄺ':('ㄹ','ㄱ'),'ㄼ':('ㄹ','ㅂ'),'ㄵ':('ㄴ','ㅈ'),'ㄱ':('','ㄱ'),'ㄷ':('','ㄷ'),'ㅂ':('','ㅂ'),'ㅈ':('','ㅈ')}
        add_dict={'ㄱ':'ㅋ','ㅈ':'ㅊ','ㄷ':'ㅌ','ㅂ':'ㅍ'}
        text=self.eleven()
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
        
        
        for i in range(l):
            if text[i][1]=='SL': continue
            #형태소내
            if len(text[i][0])>1:
                for w in range(len(text[i][0])-1):
                    #붙임1
                    if text[i][0][w][2] in ['ㄱ','ㄺ','ㄷ','ㅂ','ㄼ','ㅈ','ㄵ'] and text[i][0][w+1][0]=='ㅎ':
                        if text[i][0][w][2] =='ㄷ' and text[i][0][w+1][1]=='ㅣ':
                            text[i][0][w+1][0]=='ㅊ'
                        
                        else:
                            
                            a,b=split_dict[text[i][0][w][2]]
                          
                            text[i][0][w+1][0]=add_dict[b]
                            text[i][0][w][2]=a
                
            #형태소간
            if len(text)>1:
                
                if text[i][0][-1][2] in ['ㅎ','ㄶ','ㅀ'] and text[i+1][0][0][0] in ['ㄱ','ㄷ','ㅈ']:
                    a,b=split_dict[text[i][0][-1][2]]
                    text[i+1][0][0][0]=add_dict[text[i+1][0][0][0]]
                    text[i][0][-1][2]=a
                if text[i][0][-1][2] in ['ㅎ','ㄶ','ㅀ'] and text[i+1][0][0][0] in ['ㅅ'] :
                    
                    a,b=split_dict[text[i][0][-1][2]]
                    
                    text[i][0][-1][2]=a
                    if text[i+1][1]!='EP':
                        text[i+1][0][0][0]='ㅆ'
                if text[i][0][-1][2] in ['ㅎ'] and text[i+1][0][0][0] in ['ㄴ']:


                    text[i][0][0][2]='ㄴ'

                if text[i][0][-1][2] in ['ㅎ','ㄶ','ㅀ'] and text[i+1][0][0]=='ㅇ':

                     a,b=split_dict[text[i][0][0][2]]
                     text[i][0][-1][2]=a


                if text[i][0][-1][2] in ['ㄱ','ㄺ','ㄷ','ㅂ','ㄼ','ㅈ','ㄵ'] and text[i+1][0][0][0] =='ㅎ':
                    if text[i][1][0]=='ㄷ' or text[i+1][1]=='NNB':
                        continue

                    a,b=split_dict[text[i][0][-1][2]]
                    text[i+1][0][0][0]=add_dict[b]
                    text[i][0][-1][2]=a

            
                        
                
           
        return text
      
      
    def thirteen(self):
        thirteen_list=['ㄱ','ㄲ','ㄴ','ㄷ','ㄹ','ㅁ','ㅂ','ㅅ','ㅆ','ㅈ','ㅉ','ㅊ','ㅋ','ㅌ','ㅍ','ㅎ']
        text=self.twelve() 
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
        
        
        for i in range(l):
            #형태소 사이에서 일어나는경우  
            if len(text)>1:
            
                if (text[i][0][-1][2] in thirteen_list) and (text[i+1][1][0] in ['E','J','N','V']) and (text[i+1][0][0][0] =='ㅇ'):
                    if text[i][0][-1][2] in ['ㄷ','ㅌ'] and text[i+1][0][0][1] =='ㅣ':

                        continue
                    
                    #일연대가 이련대가 안되도록  
                    
                    if text[i+1][1][0]=='N':
                        continue
                    if text[i][1]=='NNB':
                        continue
                    if text[i+1][1]=='NNP':
                        continue

                    if text[i][1] in ['JKO','JX','ETM','MM','EC','VV']:
                        continue
                    # if text[i][1]=='NNB' and text[i+1][1]=='NNP':
                    #     continue
                    # #6월 12일 경우 6워 12릴 이된다
                    # if text[i][1]=='NNB' and text[i+1][1]=='NNB':
                    #     continue
                    # if text[i][1]=='NNP' and text[i+1][1]=='NNP':
                    #     continue
                    # if text[i][1]=='NNG' and text[i+1][1]=='NNP':
                    #     continue
                    # if text[i][1]=='NNG' and text[i+1][1]=='NNG':
                    #     continue
                    # if text[i][1][0] in ['J','M']:
                    #     continue
                # #하 ㅁ 으로써
                # if text[i][1]=='ETN' and text[i+1][1]=='JKB' and text[i+1][0][0][0]=='ㅇ':
                #     if text[i][0][0][1:]==['','']:
                #         text[i+1][0][0][0]=text[i][0][0][0]
                        

                    
                    
                    text[i+1][0][0][0]=text[i][0][-1][2]
                    text[i][0][-1][2]=''
                  #것 이 ㅁ 을
                if text[i][1]=='ETN' and text[i+1][1]=='JKO':
                    if text[i][0][-1][0]=='ㅁ' and text[i+1][0][0]==['ㅇ','ㅡ','ㄹ']:
                        text[i][0][-1] =['ㅁ','ㅡ','ㄹ']
                        text[i+1][0][0]=['','','']
                #위함 이라고
                if text[i][1]=='ETN' and text[i+1][1]=='VCP':
                    if text[i][0][-1][0]=='ㅁ' and text[i+1][0][0]==['ㅇ','ㅣ','']:
                        text[i][0][-1] =['ㅁ','ㅣ','']
                        text[i+1][0][0]=['','','']

            
            
            #형태소 안에서 일어나는경우    
            
        for i in range(len(text)):    
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in ['ㄷ','ㅌ','ㄾ'] and text[i][0][q+1][0] =='ㅇ' and text[i][0][q+1][1] =='ㅣ':
                        continue
                    if text[i][1] in ['NNP'] and text[i][0][q+1][1]!= 'ㅣ':
                        continue
                    if text[i][1] in ['JX']:
                        continue
                    if text[i][0][q][2] in thirteen_list and text[i][0][q+1][0] =='ㅇ' :             
                       
                        
                        text[i][0][q+1][0]=text[i][0][q][2]
                        text[i][0][q][2] = ''
        return text

    def fourteen(self):
        gyup_dict={'ㄳ':('ㄱ','ㅅ'),'ㄵ':('ㄴ','ㅈ'),'ㄺ':('ㄹ','ㄱ'),'ㄻ':('ㄹ','ㅁ'),'ㅄ':('ㅂ','ㅅ'),'ㄿ':('ㄹ','ㅍ'),'ㄾ':('ㄹ','ㅌ'),'ㄽ':('ㄹ','ㅅ')}
        text=self.thirteen()
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            #형태소간
            if len(text)>1:
                if (text[i][0][-1][2] in gyup_dict.keys()) and (text[i+1][1][0] in ['E','J','X']) and (text[i+1][0][0][0] =='ㅇ'):
                    a,b=gyup_dict[text[i][0][-1][2]]
                    if b=='ㅅ':
                        text[i+1][0][0][0]='ㅆ'
                        text[i][0][-1][2]=a
                    else:
                        text[i+1][0][0][0]=b
                        text[i][0][-1][2]=a
            
            if len(text[i][0])>1 and text[i][1] in ["NA",'MAG']:
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in gyup_dict.keys() and text[i][0][q+1][0] =='ㅇ':
                        a,b=gyup_dict[text[i][0][q][2]]
                        if b=='ㅅ':
                            text[i][0][q][2]=a
                            text[i][0][q+1][0]='ㅆ'
                        else:
                            text[i][0][q][2]=a
                            text[i][0][q+1][0]=b
          
        return text
    
    
    def fifteen(self):
        text=self.fourteen()
        thirteen_list=['ㄱ','ㄲ','ㄴ','ㄷ','ㄹ','ㅁ','ㅂ','ㅅ','ㅆ','ㅈ','ㅉ','ㅊ','ㅋ','ㅌ','ㅍ','ㅎ']
        nine_dict={"ㄲ":'ㄱ',"ㅋ":'ㄱ','ㅅ':'ㄷ','ㅆ':'ㄷ','ㅈ':'ㄷ','ㅌ':'ㄷ','ㅊ':'ㄷ','ㅍ':'ㅂ'}
        fifteen_dict={'ㄳ':'ㄱ','ㄺ':'ㄱ','ㅄ':'ㅂ'}
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            
            if len(text)>1:
                #형태소 간
                if (text[i+1][1][0] in ['N','V','M']) and (text[i+1][0][0][0]=='ㅇ') and (text[i+1][0][0][1] in ['ㅏ','ㅓ','ㅗ','ㅜ','ㅟ']):
                    if text[i][1][0] in ['J','E','I']:
                        continue
                    
                    if text[i+1][1] in ['NNP','NNG']:
                        continue
                   
                    if text[i][0][-1][2] in nine_dict.keys(): 
                        
                        
                        text[i+1][0][0][0]=nine_dict[text[i][0][-1][2]]
                        text[i][0][-1][2]=''
                    if text[i][0][-1][2] in ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ']:
                        text[i+1][0][0][0]=text[i][0][-1][2]
                        text[i][0][-1][2]=''
                    elif text[i][0][-1][2] in fifteen_dict.keys():
                        text[i+1][0][0][0]=fifteen_dict[text[i][0][-1][2]]
                        text[i][0][-1][2]=''
        #형태소 내         
        for i in range(len(text)):
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if (text[i][1][0] in ['N','V','M']) and (text[i][0][q+1][0]=='ㅇ') and (text[i][0][q+1][1] in ['ㅏ','ㅓ','ㅗ','ㅜ','ㅟ']):
                        if text[i][0][q][2] in thirteen_list:
                            try:
                                text[i][0][q+1][0]=self.pronounce_dict[text[i][0][q][2]]
                                text[i][0][q][2]=''
                            except:               
                                text[i][0][q+1][0]=text[i][0][q][2]             
                                text[i][0][q][2]=''   
                        
                        if text[i][0][q][2] in fifteen_dict.keys():
                            text[i][0][q+1][0]=fifteen_dict[text[i][0][q][2]]
                            text[i][0][q][2]=''
                                       
        
        
        return text
        
    def seventeen(self):
        text=self.fifteen()
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            #형태소간 변화
            if len(text)>1:
                if text[i+1][1][0] in ['J','M'] and text[i+1][0][0][0] =='ㅇ' and  text[i+1][0][0][1]=='ㅣ':
                    if text[i][0][-1][2]=='ㄷ':
                        text[i+1][0][0][0] ='ㅈ'
                        text[i][0][-1][2]=''
                    if text[i][0][-1][2]=='ㅌ':
                        text[i+1][0][0][0] ='ㅊ'
                        text[i][0][-1][2]=''
                    if text[i][0][-1][2]=='ㄾ':
                        text[i+1][0][0][0] ='ㅊ'
                        text[i][0][-1][2]='ㄹ'
            #형태소 내 변화
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2]=='ㄷ' and text[i][0][q+1][1]=='ㅣ':
                        if text[i][0][q+1][0]=='ㅇ':
                            text[i][0][q+1][0]='ㅈ'
                            text[i][0][q][2]=''
                            
                        elif text[i][0][q+1][0]=='ㅎ':
                            text[i][0][q+1][0]='ㅊ'
                            text[i][0][q][2]=''
                            
                    if text[i][0][q][2]=='ㅌ' and text[i][0][q+1][1]=='ㅣ' and text[i][0][q+1][0]=='ㅇ':
                        text[i][0][q+1][0]='ㅊ'
                        text[i][0][q][2]=''
                    if text[i][0][q][2]=='ㄾ' and text[i][0][q+1][1]=='ㅣ' and text[i][0][q+1][0]=='ㅇ':
                        text[i][0][q+1][0]='ㅊ'
                        text[i][0][q][2]='ㄹ'
                    
                           
        return text
    def eighteen(self):
        text=self.seventeen()
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
          
            #형태소간 변화
            if len(text)>1:
              
                if text[i+1][1]=='NNG' and text[i+1][1]=='VV':
                    
                    if text[i][0][-1][2] in ['ㄱ'] and text[i+1][0][0][0] in ['ㄴ','ㅁ']:
                        text[i][0][-1][2]='ㅇ'
                    if text[i][0][-1][2] in ['ㄷ'] and text[i+1][0][0][0] in ['ㄴ','ㅁ']:
                        text[i][0][-1][2]='ㄴ'
                    if text[i][0][-1][2] in ['ㅂ'] and text[i+1][0][0][0] in ['ㄴ','ㅁ']:
                        text[i][0][-1][2]='ㅁ'
            #형태소내 변화
            
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in ['ㄱ'] and text[i][0][q+1][0] in ['ㄴ','ㅁ']:
                        text[i][0][q][2]='ㅇ'
                    if text[i][0][q][2] in ['ㄷ'] and text[i][0][q+1][0] in ['ㄴ','ㅁ']:
                        text[i][0][q][2]='ㄴ'
                    if text[i][0][q][2] in ['ㅂ'] and text[i][0][q+1][0] in ['ㄴ','ㅁ']:
                        text[i][0][q][2]='ㅁ'
                    
                    
        return text   
      
    def nineteen(self):
        
        text=self.eighteen()
        if len(text)==1 and len(text[0][0])==1:
            return text
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            if len(text)>1:
                if text[i][0][-1][2] in ['ㅁ','ㅇ','ㄱ','ㅂ'] and text[i+1][0][0][0]=='ㄹ':
                    text[i+1][0][0][0]='ㄴ'
                
            if text[i][1][0] in ['N','X']:
                #형태소내 변화
                if len(text[i][0])>1:
                    for q in range(len(text[i][0])-1):
                        if text[i][0][q][2] in ['ㅁ','ㅇ','ㄱ','ㅂ'] and text[i][0][q+1][0] == 'ㄹ':
                            text[i][0][q+1][0]='ㄴ'
        
                  
        
        
        return text
      
      
    def twenty(self):
        text=self.nineteen()
        except_list1=[[['ㅇ', 'ㅣ', 'ㅁ'], ['ㅈ', 'ㅣ', 'ㄴ'], ['ㄹ', 'ㅏ', 'ㄴ']],
[['ㅅ', 'ㅐ', 'ㅇ'], ['ㅅ', 'ㅏ', 'ㄴ'], ['ㄹ', 'ㅑ', 'ㅇ']],
[['ㄱ', 'ㅕ', 'ㄹ'], ['ㄷ', 'ㅏ', 'ㄴ'], ['ㄹ', 'ㅕ', 'ㄱ']],
[['ㄱ', 'ㅗ', 'ㅇ'], ['ㄱ', 'ㅝ', 'ㄴ'], ['ㄹ', 'ㅕ', 'ㄱ']],
[['ㄷ', 'ㅗ', 'ㅇ'], ['ㅇ', 'ㅝ', 'ㄴ'], ['ㄹ', 'ㅕ', 'ㅇ']],
[['ㅅ', 'ㅏ', 'ㅇ'], ['ㄱ', 'ㅕ', 'ㄴ'], ['ㄹ', 'ㅖ', '']],
[['ㅇ', 'ㅣ', ''], ['ㅇ', 'ㅝ', 'ㄴ'], ['ㄹ', 'ㅗ', 'ㄴ']]]
        except_list2=[[['ㅇ', 'ㅢ', ''], ['ㄱ', 'ㅕ', 'ㄴ'], ['ㄹ', 'ㅏ', 'ㄴ']],[['ㅎ', 'ㅚ', 'ㅇ'], ['ㄷ', 'ㅏ', 'ㄴ'], ['ㄹ', 'ㅗ', '']],[['ㄱ', 'ㅜ', ''], ['ㄱ', 'ㅡ', 'ㄴ'], ['ㄹ', 'ㅠ', '']]]
        except_list3=[[['ㅇ', 'ㅣ', ''], ['ㅂ', 'ㅝ', 'ㄴ'], ['ㄹ', 'ㅛ', '']]]
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            #형태소간 변화
            
            if len(text)>1:
                #입원료
                if text[i][1]=='VA'and text[i+1][1]=='ETM':
                    continue
                if text[i+1][0][0][1]=='':
                    continue
                
                if len(text[i+1][0])>1:
                   
                    if [text[i][0][0],text[i+1][0][0],text[i+1][0][1]] in except_list3:
                        text[i+1][0][1][0]='ㄴ'
                
                
                #횡단로 구근류 의견란
                if len(text[i][0])>1:
                    if [text[i][0][0],text[i][0][1],text[i+1][0][0]] in except_list2:
                        text[i+1][0][0][0]='ㄴ'
                
                #길을 나섰다 -> 기를 라섣다
                if text[i][1] in ['JKO','JX']:
                    continue
                elif text[i][0][-1][2]=='ㄴ' and text[i+1][0][0][0]=='ㄹ':
                    text[i][0][-1][2]='ㄹ'
                elif text[i][0][-1][2]=='ㄹ' and text[i+1][0][0][0]=='ㄴ':
                    text[i+1][0][0][0]='ㄹ'
                elif text[i][0][-1][2]=='ㅀ' and text[i+1][0][0][0]=='ㄴ':
                    text[i+1][0][0][0]='ㄹ'
                    text[i][0][-1][2]='ㄹ'
                elif text[i][0][-1][2]=='ㄾ' and text[i+1][0][0][0]=='ㄴ':
                    text[i+1][0][0][0]='ㄹ'
                    text[i][0][-1][2]='ㄹ'
                          
                    
             #형태소 내 변화
        for i in range(len(text)):    
            
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][0] in except_list1:
                        if text[i][0][q][2] =='ㄴ' and text[i][0][q+1][0]=='ㄹ':
                            text[i][0][q][2]='ㄴ'
                        if text[i][0][q][2] =='ㄹ' and text[i][0][q+1][0]=='ㄴ':
                            text[i][0][q+1][0]='ㄴ'
                        
                    if text[i][0][q][2] =='ㄴ' and text[i][0][q+1][0]=='ㄹ':
                        
                        text[i][0][q][2]='ㄹ'
                    if text[i][0][q][2] =='ㄹ' and text[i][0][q+1][0]=='ㄴ':
                        text[i][0][q+1][0]='ㄹ'
        return text
    def t_three(self):
        text=self.twenty()
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
                                                             
            
              #형태소 내에서 
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][0][q][2] in ['ㄱ','ㄷ','ㅂ'] and text[i][0][q+1][0] in self.dsr_dict.keys():
                        text[i][0][q+1][0]=self.dsr_dict[text[i][0][q+1][0]]
            
            #형태소간 일어나는
            if len(text)>1:
                if text[i+1][1][0] in ['M','N','N','X']:
                    continue
                if text[i][1][0] in ['M','X']:
                    continue
               
                if text[i][0][-1][2] in ['ㄱ','ㄷ','ㅂ','ㅆ'] and text[i+1][0][0][0] in self.dsr_dict.keys():
                    text[i+1][0][0][0]=self.dsr_dict[text[i+1][0][0][0]]
          
        return text
    def t_four(self):
        text=self.t_three()
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            
            
            
            if len(text)>1:
                if text[i][1][0]=='V' and text[i][0][-1] in [['ㄱ','ㅣ',''],['ㄷ','ㅗ','ㄴ']] and text[i+1][0][0]==['ㄷ','ㅏ','']:
                    continue
                
                
                if text[i][1][0]=='V' and text[i][0][-1][2] in ['ㄴ','ㅁ']:
                    if text[i+1][1][0]=='E' and text[i+1][0][0][0] in ['ㄱ','ㄷ','ㅅ','ㅈ']:
                        text[i+1][0][0][0]=self.dsr_dict[text[i+1][0][0][0]]
                      
        return text
    
    
    def t_five(self):
        text=self.t_four()
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
                     
            if len(text)>1:
                if text[i][1][0]=='V' and text[i][0][-1][2] in ['ㄼ', 'ㄾ']:
                    
                    if text[i+1][1][0]=='E' and text[i+1][0][0][0] in ['ㄱ', 'ㄷ', 'ㅅ', 'ㅈ']:
                        text[i+1][0][0][0]=self.dsr_dict[text[i+1][0][0][0]]
                
        return text 
      
      
    def t_six(self):
        text=self.t_five()
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][1]=='NNG' and text[i][0][q][2] =='ㄹ' and text[i][0][q+1][0] in ['ㄷ','ㅅ','ㅈ']:
                        if text[i][0][q]==text[i][0][q+1]:
                            continue
                        else:
                            text[i][0][q+1][0]=self.dsr_dict[text[i][0][q+1][0]]
        return text                    
          
    def t_seven(self):
      
        text=self.t_six()
        
        if len(text)==1:
            l=1
        else:
            l=len(text)-1
            
        for i in range(l):
            if len(text)>1:
                if text[i][1]=='ETM' and text[i][0][0]==['ㄹ','',''] and text[i+1][0][0][0] in ['ㄱ', 'ㄷ', 'ㅂ', 'ㅅ', 'ㅈ']:
                    text[i+1][0][0][0]=self.dsr_dict[text[i+1][0][0][0]]
            
        
        for i in range(len(text)):    
            if len(text[i][0])>1:
                for q in range(len(text[i][0])-1):
                    if text[i][1]=='EC' and text[i][0][q]==['ㄹ','',''] and text[i][0][q+1][0] in ['ㄱ', 'ㄷ', 'ㅂ', 'ㅅ', 'ㅈ']:
                        text[i][0][q+1][0]=self.dsr_dict[text[i][0][q+1][0]]


        
        return text
      
    def t_nine(self):
        text=self.t_seven()
        
        
        for i in range(len(text)):
            if len(text[i][0])>1 and len(text[i])==3:   #Compund인경우만 len(text[i]) 가 3이고 나머진2
                for q in range(len(text[i][0])-1):
                                                              
                    if text[i][0][q][2] not in [''] and text[i][0][q+1][0]=='ㅇ' and text[i][0][q+1][1] in ['ㅣ','ㅑ','ㅕ','ㅛ','ㅠ']:
                        if text[i][0][q][2]=='ㄹ':
                            text[i][0][q+1][0]='ㄹ'    
                        else:
                            text[i][0][q+1][0]='ㄴ'
        return text

         
    
                        
        
    def thirty(self):
        text=self.t_nine()
        text2=self.listed_text
        #형태소 내 변화
        for i in range(len(text)):
            if len(text2[i][0])>1:
                for q in range(len(text2[i][0])-1):
                    if text2[i][1] in ['NNG','NA'] and text2[i][0][q][2] =='ㅅ': 
                        if text2[i][0][q+1][0] in ['ㄱ', 'ㄷ', 'ㅂ', 'ㅅ', 'ㅈ'] :
                            text[i][0][q+1][0]=self.dsr_dict[text2[i][0][q+1][0]]
                        if text2[i][0][q+1][0] in ['ㄴ','ㅁ']:
                            text[i][0][q][0]='ㄴ'
                        if text2[i][0][q+1][0]=='ㅇ' and text2[i][0][q+1][1]=='ㅣ':
                            text[i][0][q][2]='ㄴ'
                            text[i][0][q+1][0]='ㄴ'
        #형태소간
        if len(text[i][0])>1:
            for i in range(len(text2)-1):
                if text2[i][1]=='NNG' and text2[i+1][1]=='NNP':
                    if text2[i+1][0][0][0] in ['ㄱ', 'ㄷ', 'ㅂ', 'ㅅ', 'ㅈ'] and text2[i][0][-1][2]=='ㅅ' :
                        text2[i+1][0][0][0]=self.dsr_dict[text2[i+1][0][0][0]]
                       
        
        return text     
    
    
    def make_sentence(self):
        from hgtk.letter import compose,decompose
        if self.listed_text==[]:
            text=[[[['', '', '']], '']]
        else:    
            text=self.thirty()
        composed=''
        
        for i in range(len(text)-1):
            
#             if text[i][0][0]==['ㅎ','ㅡ',''] and text[i][0][1]==['ㄹ','ㅡ','']:
#                 if text[i+1][0][0][0]!='ㅁ':
#                     text[i][0]=[['ㅎ','ㅡ','ㄹ']]
#                     text[i+1][0][0][0]='ㄹ'
#             if text[i][0][-1]==['ㄸ','ㅐ',''] and text[i+1][0][0]==['ㅇ','ㅣ','']:
#                 if text[i+1][1]=='VCP' and text[i][1]=='NNG':
#                     text[i+1][0][0]=['','','']
            
#             if text[i][0][-1][-1] in ['ㅂ','ㅁ'] and text[i+1][0][0]==['ㄴ','',''] :
#                 text[i][0][-1][2]=''
#                 text[i+1][0][0]=['ㅇ','ㅜ','ㄴ']
#             elif text[i+1][0][0][1]=='' and text[i+1][0][0][0]!='' :
#                 text[i][0][-1][2]=text[i+1][0][0][0]
# #                 text[i+1][0][0]=['','','']
            
            for q in (text[i][0]):

                try:
                    composed+=compose(*q)
                except:
                    None
        
        
        
        for w in text[-1][0]:

            try:
                composed+=compose(*w)
            except:
                None
        text_=[r for r in composed]
        list(map((lambda x: text_.insert(*x)),self.ww))
       
        composed=''.join(text_)
 
        return composed
    



def translate_text():
    text=input('type text:')
    text=text.replace('"',"'")
    w=prepro(text)
    return w.make_sentence()

def translate_line():
    text=input('type text:')
    textt=text.split('.')
    textt=textt[:-1]
    text=[]
    for i in textt:
        text.append(i)
        print(i)
        text.append(prepro(i).make_sentence())
        text.append(' ')
    return text
    
    

```
