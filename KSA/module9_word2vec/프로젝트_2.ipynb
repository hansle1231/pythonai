```python
import re
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
#cv = CountVectorizer(max_features=1000, tokenizer=get_nouns)
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from konlpy.tag import Okt
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk
from konlpy.corpus import kobill
from konlpy.tag import Hannanum
import gensim
import sklearn.decomposition
han=Hannanum()
okt=Okt()
```

```python
article_list2 = pd.read_csv(r'C:\Users\han\Downloads\new_crawl.csv', encoding='utf-16')

article_list2.head()
```

```python
a1 = article_list2.values.tolist()

a1
```

```python
# 클리닝 함수
def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]', ' ',str(text))
    cleaned_text = re.sub('[→…★\{\}\[\]新\/?.,;:⑦|\♥)*~`·!^\-_+<>@\#$%&\\\=\(\'\"]',
                          ' ', cleaned_text)
    cleaned_text = re.sub('[0-9]', ' ',cleaned_text)
    return cleaned_text

list_pr=[]
for i in np.arange(article_list2.shape[0]):
    temp=a1[i]
    list_pr.append(temp)
print(list_pr)
```

```python
article_flat = []
for item in list_pr:
    article_flat.extend(item)

article_flat
```

```python
text=clean_text(article_flat)

text
```

```python
text
```

```python
words=okt.pos(text)

from gensim.models import Word2Vec
model = Word2Vec(sentences=words, size=100, window=5, min_count=1, workers=4, sg=0) ## 우리가 크롤링한 자료로 word2vec training

word_vectors = model.wv
word_vectors
vocabs = word_vectors.vocab.keys()
vocabs
```

```python
similar=model.wv.most_similar("연습실")
similar
```
