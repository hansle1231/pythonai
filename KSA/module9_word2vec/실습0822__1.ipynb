```python
# 클리닝 함수
def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]', ' ',str(text))
    cleaned_text = re.sub('[→…★\{\}\[\]新\/?.,;:⑦|\♥)*~`·!^\-_+<>@\#$%&\\\=\(\'\"]',
                          ' ', cleaned_text)
    cleaned_text = re.sub('[0-9]', ' ',cleaned_text)
    return cleaned_text
```

# 문장토큰화 - > 단어토큰화

```python
#문장토큰화

senten_tok=[]
for i in list_pr:
    senten_tok+=(i[0].split('.'))
```

```python
senten_tok[0]
```

### 각문장을 단어토큰화

```python
han=Hannanum()
han.pos(senten_tok[1])
```

### 한나눔이 명사분류를 더잘하는것같다.

```python
okt.pos(senten_tok[1])
```

```python
result=[]
for i in senten_tok:
    temp=[]
    a=han.pos(i)
    for j in a:
        if j[1]=='N':
            temp.append(j[0])
    result.append(temp)
        
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences=result, size=100, window=5, min_count=1, workers=4, sg=0) ## 우리가 크롤링한 자료로 word2vec training
```

```python
word_vectors = model.wv
word_vectors
vocabs = word_vectors.vocab.keys()
vocabs
```
