```python
import urllib.request
import urllib.parse

API='http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
#루트주소,, 물음표앞주소

val={'.stnId':'108'}
params=urllib.parse.urlencode(val)  #한글일떄 인코드(커리문에 한글들어가면 안되는경우가 대부분)
#params:'stnId=108'
print(params)


#요청 url
url=API+'?'+params
print('url={0}'.format(url))

#다운로드

res=urllib.request.urlopen(url)
data=res.read()   #바이너리형식으로 읽어옴
text=data.decode('utf-8') #디코드해서 텍스트형식으로
print(text)
```

```python
import sys
import urllib.request as req
import urllib.parse as parse

text=[]
while (True):
    regionNumber=input('ID:')
    
    if regionNumber.upper()=='EXIT':
        break
    elif (int(regionNumber) not in [108,109,105,131,133,146,156,134,159,184]):
            continue
            
    API='http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?'
    values={'stnId':regionNumber}
    params=parse.urlencode(values)
    url=API+params
    print('url=',url)
    data=req.urlopen(url).read()
    text.append(data.decode('utf-8'))
```
