```python
import codecs
from bs4 import BeautifulSoup
from konlpy.tag import Twitter
```

```python
fp=codecs.open(r'C:\Users\Affinity\Downloads\modeule04\ch02\BEXX0003.txt','r',encoding='utf-16')
```

```python
soup=BeautifulSoup(fp,'html.parser')
body=soup.select_one('body>text')
text=body.getText()
```

```python
print(text)
```

```python
twitter=Twitter()
word_dic={}
lines=text.split('\n')
for line in lines:
    mallst=twitter.pos(line)  #문장마다 형태소분석
    for word in mallst:
        if word[1]=='Noun': #명사확인
            if not word[0] in word_dic :#단어 :빈도   몇개나오는지 체크
                word_dic[word[0]]=0
            
            word_dic[word[0]] += 1
        
keys=sorted(word_dic.items(),key=lambda x : x[1],reverse=True)
for word, count in keys[:50]:
    print('{0}{1}'.format(word,count),end='\n')
print()
    
    
```
